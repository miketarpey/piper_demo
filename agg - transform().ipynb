# Setup

```python
from piper.defaults import *
```

# Sample data

```python
df = sample_data()
head(df, 10)
```

# Examples

## groupby

```python
%%piper

df.groupby(['countries', 'regions']).agg(TotSales1=('values_1', 'sum'))
>> transform()
>> head(8)
>> adorn()
```

## count with add_total option

```python
count(df, 'countries', totals=True)
```

## counts() - multindex total with adorn()

```python
%%piper
count(df, ['regions', 'countries'], sort_values=None)
>> order_by(['regions', '-n'])
>> reset_index()
>> head(4)
>> adorn(ignore_index=True)
```

## counts() - multindex total with adorn()

```python
%%piper

df.groupby(['countries', 'regions']).agg(total=('values_1', 'sum'))
>> where("countries.isin(['France', 'Italy', 'Germany'])")
>> transform()
>> adorn(axis='row', ignore_index=False)
>> adorn(axis='column', columns='total')
```

```python
gx = pivot_table(df, index=['countries', 'regions'], freq='A', 
                 format_date=True, values=['values_1', 'values_2'])
gx = gx.unstack().fillna(0).astype(int)
gx = flatten_names(gx, remove_prefix='values_1_')
gx = flatten_names(gx, remove_prefix='values_2_')
gx = gx.reset_index()
# gx = adorn(gx)
gx
```
