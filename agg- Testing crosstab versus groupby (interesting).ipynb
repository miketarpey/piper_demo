# Setup

```python
from piper import piper
from piper.verbs import *
from piper.factory import get_sample_data
```

```python
df = get_sample_data()
head(df)
```

```python
# ordered_category = ['North', 'South', 'East', 'West']
# ordered_category = pd.CategoricalDtype(ordered_category)
# df.regions = df.regions.astype(ordered_category)
```

## Normal usage (count())

```python
%%piper
pd.crosstab(df.countries, df.regions, margins=True)
```

```python
%%piper
df.groupby(['countries', 'regions'])['values_1'].count().unstack()
>> adorn(axis='both')
```

# using sum of values

```python
%%piper
pd.crosstab(df.countries, df.regions, values=df.values_1,
            aggfunc='sum', margins=True)
```

```python
%%piper
df.groupby(['countries', 'regions'])['values_1'].sum().unstack()
>> adorn(axis='both')
```

## normalize data

```python
# Doesn't work if regions are defined as a categorical
# TypeError: cannot append a non-category item to a CategoricalIndex
```

```python
%%piper
pd.crosstab(df.countries, df.regions, values=df.values_1,
            aggfunc='sum', margins=False, normalize='index').round(5)
>> adorn(axis='column')
```

```python
%%piper
pd.crosstab(df.countries, df.regions, values=df.values_1,
            aggfunc='sum', margins=False, normalize='columns').round(6)
>> adorn(axis='row')
```
