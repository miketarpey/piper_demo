```python
import requests
from bs4 import BeautifulSoup

url = 'https://freebiesglobal.com/'
url = 'https://freebiesglobal.com/dealstore/udemy'
# url = 'https://freebiesglobal.com/dealstore/amazon'
url = 'https://freebiesglobal.com/best-sellers'
url = 'https://freebiesglobal.com/dealstore/udemy'
response = requests.get(url)

status = response.status_code 
json = response.text

soup = BeautifulSoup(json, 'lxml')
```

```python
# find pagination information
pagination = soup.find('div', class_='pagination').find('ul').find_all('li')[4]
pagination_link = pagination.find('a')['href']
total_pages = int(pagination.text)
total_pages
```

```python
courses = []

articles = soup.find_all('article', class_="col_item offer_grid offer_grid_com mobile_compact_grid")

for article in articles:

    deal_meta = article.find('div', class_="meta_for_grid")
#     print(deal_meta)
    
    category = deal_meta.find('a', class_="cat").text
#     print(category)
    
    deal = article.find('div', class_="info_in_dealgrid")
#     print(deal)
  
    discount = deal.find('span', class_='grid_onsale')
    if discount:
        discount = discount.text
        
    actual_price = deal.find('span', class_='rh_regular_price')
    if actual_price:
        actual_price = actual_price.text
        
    original_price = deal.find('span', class_='rh_price_wrapper').find('del')
    if original_price:
        original_price = original_price.text
#     print(regular_price)
    
    source = deal.find_all('a')[1].find('img')['alt']
#     print(source)
    
    deal_details = deal.find('a', class_='img-centered-flex rh-flex-center-align rh-flex-justify-center')
#     print(deal_details)
    course_title = deal_details.find('img')['alt']
    course_link = deal_details['href']
    
    # Expired offer?
    try:
        expired = deal.find('span', class_="rh-expired-notice").text
    except AttributeError as e:
        expired = ''
#     print(expired)
    
    course_data = {'category': category,
                   'source': source,
                   'title': course_title, 
                   'course_link': course_link,
#                    'original_price': original_price,
                   'expired': expired,
                   'discount': discount, 
                   'actual_price': actual_price
                  }
    
    courses.append(course_data)
    
#     break
```
